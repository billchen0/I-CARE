```python
%reload_ext autoreload
%autoreload 2
from pathlib import Path
import neurokit2 as nk
import numpy as np
import sys
sys.path.append("..")
from feature_extraction import extract_features

root_path = Path("/media/hdd1/i-care/processed/")
```

```python
# Select one patient to construct the pipeline
p_data = root_path / "0326"
```

```python
# Loop through the hourly segments of EEG data
for hr_seg in sorted(p_data.iterdir()):
    # Read in the segment
    eeg_data = np.load(hr_seg)
    total_samp = eeg_data.shape[0]
    window_samp = 5*60*100
    # Loop through each segment with a 5 minute window
    for start in range(0, eeg_data.shape[0], window_samp):
        end = start + window_samp
        # Drop segments that don't have full length
        if end <= total_samp:
            segment = eeg_data[start:end, :]
        # Extract features from the segment
        feature = extract_features(segment)
        print(feature)
```
